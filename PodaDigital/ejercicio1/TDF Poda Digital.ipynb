```python
df1=pd.read_csv ("tdf_winners.csv")
```

```python
df1["speed"]=df1["distance"]/df1["time_overall"]
```

```python
df1["year"]=df1["start_date"].apply(lambda x: x[0:4])
df1.head()
```

```python
df1[df1['year']>1919].head(30)
```

```python
df1['year'] = df1['year'].astype(int)
```

```python
import matplotlib.pyplot as plt
y=df1['speed']
x=df1['year']


fig, ax = plt.subplots(figsize=(15, 8))


ax.scatter(x,y, color='blue')

fig.set_facecolor('xkcd:white')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
ax.set_xlabel('year', fontsize=20)
ax.set_ylabel('average speed', fontsize=20)
#ax.set_title('Average speed of winners', color="red", fontsize = 20)
ax.grid()


plt.savefig('speed.png')
plt.show()
```
